С обучением модели на датасете из тьюторила kaggle у меня возникли проблемы (я писала в письме, что при обучении модель ограничивается только 27 словами -- в логе все время пишется "keeping 27 word types"), поэтому я решила обучить модель на другом доступном английском корпусе. 

Я скачала корпус в 1 млрд слов, который доступен по адресу http://code.google.com/p/1-billion-word-language-modeling-benchmark/. Корпус состоит из текстов, в которых все токены, включая пунктуацию, разделены пробелом. Поскольку корпус большой, обрабатывая исходные тексты (т.е. разделяя тексты на предложения и удаляя знаки препинания), я разделила весь массив текстов на 4 части и далее последовательно обучала модель. 
На первой части модель обучилась за 2,5 часа, но дообучения второй частью я не дождалась и разделила исходный корпус на 8, а потом и на 12 частей. Модель стала дообучаться быстрее, но и при 8, и при 12 частях наступал момент (на этапе "matching count from corpus used for vocabulary survey"), когда я не могла дождаться окончания дообучения (у меня один и не очень мощный ноутбук).

(сначала не сообразила и по привычке все делала в pycharm'е, а перезапускать код в ноутбуке долго)

К сожалению, на это ушло много времени, и я уже не успевала скачать и обработать какой-нибудь другой корпус, поэтому взяла брауновский корпус из nltk (хотя он, конечно, не подходит для этой задачи -- всего 57340 предложений и чуть более миллиона токенов).

```python
import logging
import gensim
from nltk.corpus import brown
from string import punctuation

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
logging.root.level = logging.INFO

punct = set(punctuation)          
sentences = []
words = 0
for sent in brown.sents():
    words += len(sent)
    sent = [w.lower() for w in sent if w not in punct]
    sentences.append(sent)
print('Количество предложений:', +len(sentences))
print('Количество слов:', +words)
```

При обучении модели использовала те же параметры, что в тьюториле kaggle.

```python
model = gensim.models.Word2Vec(sentences, workers=4, size = 300, min_count = 40, window = 10, sample = 1e-3)
```

Корпус маленький, обучается быстро (9,2с). Но и точность на тестовых запросах показывает маленькую -- 4%.

```python
model.accuracy('/home/anya/word2vec_assignment/questions.txt')
```

Брауновский корус более менее сбалансирован по жанрам и типам текста, поэтомы было бы справедливо ожидать хорошей модели, если бы не размер. Сбалансированность корпуса может оказаться значимее, чем размер (так, модель на НКРЯ более репрезентативна (лучше "отражает действительность"), чем модель на значительно большем веб-корпусе), но все же 1 млн. токенов мало.
Если не убирать из модели наиболее частотные слова (хотя обычно суб-сэмплирование только на пользу), поднять порог минимальной частоты с 40 до 55, уменьшить окно контекста с 10 до 5, то точность несколько повышается: 8,3%.

```python
model = gensim.models.Word2Vec(sentences, workers=4, size = 300, min_count = 55, window = 5, sample = 0)
```

```python
model.accuracy('/home/anya/word2vec_assignment/questions.txt')
```

Можно предположить, что ошибки модели, обученной на предоставленных kaggle данных, будут в больше степени объясняться спецификой текстов -- ревью на фильмы, и лишь отчасти размерами корпуса (точности в 70%, как на гугловском 1,6 млрд. датасете, конечно, не будет, но около 40%, мне кажется, будет). В то время как причина ошибок модели, обученной на брауновском корпусе, это в первую очередь его размер и, соответственно, маленький словарь модели.

Примеры на поиск лишних слов было довольно сложно составить так, чтобы все четыре слова имелись в словаре модели.

Например в этом примере может показаться, что модель ошибается, но на самом деле ни lunch, ни cereal, ни breakfast просто нет в словаре модели:

```python
model.doesnt_match("breakfast cereal dinner lunch".split())
```

```python
model.doesnt_match("school child children office".split())
```

Модель ошибается, но вполне логично, что в контексте школы в корпусе заметно больше говорят про детей, а не про ребенка. 

```python
model.doesnt_match("school parents children office".split())
```

Теперь все нормально.

```python
model.doesnt_match("price home cost shop".split())
```

Всё верно!

```python
model.doesnt_match("green heavy blue yellow".split())
```

Всё верно!

```python
model.doesnt_match("sleep talk speak ask".split())
```

Всё верно, глаголы talk, speak, ask достаточно частотные, поэтому их контексты соответствуют здравому смыслу (в предыдущих двух примерах такая же ситуация).

```python
model.doesnt_match("rain sun snow wind".split())
```

Возможно, это произошло чисто случайно, но интересно, что модель разделяет rain, snow, wind и sun как слова, для которых характерны негативные и позитивные контексты соответственно.
